```python
import pandas as pd
from itertools import combinations
import ast
import numpy as np
import warnings
warnings.filterwarnings('ignore')
pd.set_option('display.max_columns', None)
```

```python
data = pd.read_csv("data_final.csv")
```

```python
# Let's write a function that processes a list of genre strings and constructs the DataFrame.

from itertools import combinations
import ast

# Sample data (a list containing multiple strings like the given example)
genre_strings = data['genres'].tolist()

# Initialize an empty DataFrame for the edge list
edges_df = pd.DataFrame(columns=['genre_A', 'genre_B', 'weight'])

# Process each genre string in the list
for genre_str in genre_strings:
    # Convert the string representation of list to an actual list of dictionaries
    try:
        genres = ast.literal_eval(genre_str)
    except Exception as e:
        print(e)
        genres = []
        
    
    # Extract the 'description' of each genre
    descriptions = [genre['description'] for genre in genres]
    
    # Generate all unique pairs of genres and increment their weight
    for genre_pair in combinations(descriptions, 2):
        # Sort the genres alphabetically to ensure unique representation
        genre_pair = tuple(sorted(genre_pair))
        
        # Skip self loops
        if genre_pair[0] == genre_pair[1]:
            continue
        
        # Check if this pair is already in the DataFrame
        if genre_pair in edges_df.set_index(['genre_A', 'genre_B']).index:
            # Increment the weight for this pair
            edges_df.loc[(edges_df['genre_A'] == genre_pair[0]) & (edges_df['genre_B'] == genre_pair[1]), 'weight'] += 1
        else:
            # Add the new pair to the DataFrame with a weight of 1
            edges_df = edges_df.append({'genre_A': genre_pair[0], 'genre_B': genre_pair[1], 'weight': 1}, ignore_index=True)

# Ensure the 'weight' column is of type int
edges_df['weight'] = edges_df['weight'].astype(int)

edges_df  # Display the edge DataFrame

```

```python
edges_df = edges_df.sort_values(by='weight', ascending=False)
```

```python
def helper(year):
    data = pd.read_csv(f'data_{year}_newly_added.csv')
    genre_col = data['genres'].tolist()
    genre_list = ast.literal_eval(genre_col[0])
    genres = [genre['description'] for genre in genre_list]


    # Sample data (a list containing multiple strings like the given example)
    genre_strings = genre_col

    # Initialize an empty DataFrame for the edge list
    edges_df = pd.DataFrame(columns=['genre_A', 'genre_B', 'weight'])

    # Process each genre string in the list
    for genre_str in genre_strings:
        # Convert the string representation of list to an actual list of dictionaries
        try:
            genres = ast.literal_eval(genre_str)
        except Exception as e:
#             print(e)
            genres = []


        # Extract the 'description' of each genre
        descriptions = [genre['description'] for genre in genres]

        # Generate all unique pairs of genres and increment their weight
        for genre_pair in combinations(descriptions, 2):
            # Sort the genres alphabetically to ensure unique representation
            genre_pair = tuple(sorted(genre_pair))

            # Skip self loops
            if genre_pair[0] == genre_pair[1]:
                continue

            # Check if this pair is already in the DataFrame
            if genre_pair in edges_df.set_index(['genre_A', 'genre_B']).index:
                # Increment the weight for this pair
                edges_df.loc[(edges_df['genre_A'] == genre_pair[0]) & (edges_df['genre_B'] == genre_pair[1]), 'weight'] += 1
            else:
                # Add the new pair to the DataFrame with a weight of 1
                edges_df = edges_df.append({'genre_A': genre_pair[0], 'genre_B': genre_pair[1], 'weight': 1}, ignore_index=True)

    # Ensure the 'weight' column is of type int
    edges_df['weight'] = edges_df['weight'].astype(int)

    edges_df = edges_df.sort_values(by='weight', ascending=False)
    
    return edges_df
```

```python
# Exclude rows with a '0' year
filtered_data = data[data['year'] != '0']

print(filtered_data.shape)
# Get unique years
unique_years = filtered_data['year'].unique()

# Save each year's data to a separate CSV file
for year in unique_years:
    print(year)
    year_data = filtered_data[filtered_data['year'] == year]
    print(year_data.shape)
#     print(year_data.head())
    year_data.to_csv(f'data_{year}_newly_added.csv', index=False)  
```
